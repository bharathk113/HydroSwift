# SWIFT CLI Examples Notebook

This notebook mirrors the workflows shown in `PYTHON_API_EXAMPLES.ipynb`, but uses the command-line interface instead of the Python API. Run these cells from the repository root or from an environment where the `swift` command is available.


## Getting the WRIS data

This can be done directly from the CLI by specifying a basin and one or more WRIS variable flags.


### Discover available basins and CWC coverage

The CLI does not return filterable tables like the Python API, so a typical CLI workflow starts by listing available basins/station coverage and then issuing explicit download commands for the basins or stations you want.


```python
!swift --list
```

### WRIS time-series download for selected basins and variables

This is the CLI equivalent of selecting a subset of WRIS basins in the Python notebook and then calling `swift.fetch(...)` or `swift.wris.download(...)`.


```python
%%bash
swift -b Cauvery -sed -wl \
  --output-dir data_fetch_wris \
  --start-date 2024-04-01 \
  --end-date 2024-06-07 \
  --merge \
  --plot \
  --overwrite

swift -b Godavari -sed -wl \
  --output-dir data_fetch_wris \
  --start-date 2024-04-01 \
  --end-date 2024-06-07 \
  --merge \
  --plot \
  --overwrite
```

### WRIS download using station-like explicit selections

The Python notebook uses `wris.stations(...)` and then passes the resulting table into `fetch(...)`. In the CLI, the nearest equivalent is to run direct basin + variable commands for the combinations you want to download.


```python
%%bash
swift -b Godavari -solar -sed \
  --output-dir data_fetch_wris \
  --start-date 2024-04-01 \
  --end-date 2025-05-07 \
  --merge \
  --plot \
  --overwrite

swift -b Narmada -solar -sed \
  --output-dir data_fetch_wris \
  --start-date 2024-04-01 \
  --end-date 2025-05-07 \
  --merge \
  --plot \
  --overwrite
```

### WRIS download with a different basin/variable combination


```python
%%bash
swift -b Krishna -q -solar \
  --output-dir data_download_wris \
  --start-date 2024-01-01 \
  --end-date 2024-02-07 \
  --merge \
  --plot \
  --overwrite

swift -b Tapi -q -solar \
  --output-dir data_download_wris \
  --start-date 2024-01-01 \
  --end-date 2024-02-07 \
  --merge \
  --plot \
  --overwrite
```

# Getting the CWC data

CWC provides near-real-time water-level data. In the CLI you can download either by basin or by explicit station code.


### CWC time-series download by basin

This mirrors the Python notebook examples that start from `swift.cwc.basins()` and then fetch a subset of basins.


```python
%%bash
swift --cwc-basin Narmada Tapi \
  --output-dir data_fetch_cwc \
  --start-date 2024-04-01 \
  --end-date 2024-06-07 \
  --merge \
  --plot \
  --overwrite
```

#### CWC station download by explicit station code

This is the CLI equivalent of filtering a `swift.cwc.stations(...)` result table and then downloading only those stations.


```python
%%bash
swift --station 040-CDJAPR 032-LGDHYD \
  --output-dir data_fetch_cwc \
  --start-date 2024-04-01 \
  --end-date 2024-06-07 \
  --merge \
  --plot \
  --overwrite
```

### CWC download with another basin selection


```python
%%bash
swift --cwc-basin Narmada Tapi \
  --start-date 2024-01-01 \
  --end-date 2024-01-07 \
  --merge \
  --plot \
  --overwrite
```

```python
!swift --merge-only --input-dir data_fetch_cwc --output-dir merged_cwc --cwc
```

If you download with `--plot` disabled, you can later generate plots from existing files using `--plot-only`.


```python
!swift --plot-only --input-dir data_fetch_cwc --output-dir plots_cwc --cwc --plot-svg --plot-moving-average-window 30
```
