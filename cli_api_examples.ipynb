## ⚙️ Setup (IMPORTANT)

If you see `swift: command not found`, install the package in editable mode:

```bash
pip install -e .
```

Ensure your CLI entry point is defined in `pyproject.toml`:

```toml
[project.scripts]
swift = "swift_app.__main__:main"
```

If using a virtual environment:

```bash
source .venv/bin/activate
```

Fallback (always works):

```bash
python -m swift_app --help
```
