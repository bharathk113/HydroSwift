## Getting the WRIS data
This can be done either using the generic fetch() method or the wris.download() method.


### WRIS time series data download using fetch()

```python
# The list of basins or stations can be queried from the WRIS database using the wris.basins() or wris.stations() methods.
# This list can then be passed on to the fetch() to download the time series data.
basins_wris = swift.wris.basins(variable=["sediment", 'water_level']) # Returns a pandas Dataframe with the basins and their WRIS IDs for sediment data. You can specify other variables as needed.
basins_wris
```

```python
# If needed, we can subset the returned dataframe containing the basins before downloading the data. 
# For example, if we are only interested in the sediment and water level data for the basins Cauvery and Godavari, 
# we can filter the dataframe as follows:
basins_wris_subset = basins_wris[basins_wris['basin'].isin(['Cauvery', 'Godavari'])]
basins_wris_subset
```

```python
# Then we pass this subset table to the fetch function to download the data for these basins.
# Test fetch() from CWC basin table

gdf_wris = swift.fetch(
    basins_wris_subset,# the subset dataframe containing the basins and their WRIS IDs for sediment and water level data.
    output_dir="data_fetch_wris", #location where the downloaded files will be saved.
    start_date="2024-04-01",
    end_date="2024-06-07",
    merge=True, # merges and returns all the downloaded files as a geodataframe.
    plot=True, # plots the station-wise time series of the downloaded data.
    quiet=False, # prints the progress of the download and any warnings or errors encountered during the download process.
    overwrite=True # overwrites any existing files in the output directory with the same name as the downloaded files.
)

# gdf_wris will contain the merged geodataframe of the downloaded data for the specified basins and variables.
```

```python
# gdf_wris will contain the merged geodataframe of the downloaded data for the specified basins and variables.
gdf_wris
```

We can also use the fetch () function to donwload the WRIS time series data by passing the stations table returned from wris.stations()

```python
stations = swift.wris.stations(basin=["Godavari", "Narmada"], variable=["solar", "sediment"])
stations
```

```python
# fetch() also works with WRIS station table; if for the requested time period,
# the data for a station is not available, it will be skipped 
# and the download will continue for the other stations.
gdf_wris_1 =swift.fetch(
    stations,
    output_dir="data_fetch_wris",
    start_date="2024-04-01",
    end_date="2025-05-07",
    merge=True,
    plot=True,
    quiet=False, overwrite=True
)
gdf_wris_1 
# The returned geodataframe (gdf_wris_1 in this case) will contain the merged data 
# for the stations and variables for which data was available for the specified time period.
```

# Getting the CWC data
- This can be done either using the generic fetch() method or the cwc.download() method.

- CWC only provides the near real-time water level data at gauge stations


### CWC time series data (only water level) using fetch()
- The list of basins or stations can be queried from the CWC database using the cwc.basins() or cwc.stations() methods.

- This list can then be passed on to the fetch() to get the water level time series data.

```python

basins_cwc = swift.cwc.basins() # Returns a pandas Dataframe with the basins and their CWC IDs.
basins_cwc
```

```python
# If needed, we can subset the returned dataframe containing the basins before downloading the data. 
# For example, if we are only interested in the sediment and water level data for the basins Cauvery and Godavari, 
# we can filter the dataframe as follows:
basins_cwc_subset = basins_cwc[basins_cwc['basin'].isin(['Narmada', 'Tapi'])]
basins_cwc_subset
```

```python
# Then we pass this subset table to the fetch function to download the data for these basins.

gdf_cwc = swift.fetch(
    basins_cwc_subset,# the subset dataframe containing the basins and their CWC IDs.
    output_dir="data_fetch_cwc", #location where the downloaded files will be saved.
    start_date="2024-04-01",
    end_date="2024-06-07",
    merge=True, # merges and returns all the downloaded files as a geodataframe.
    plot=True, # plots the station-wise time series of the downloaded data.
    quiet=False, # prints the progress of the download and any warnings or errors encountered during the download process.
    overwrite=True # overwrites any existing files in the output directory with the same name as the downloaded files.
)

# gdf_cwc will contain the merged geodataframe of the downloaded data for the specified basins and variables.
```

#### We can also query the available stations filtered by basin and state names using cwc.stations()

```python
stations_cwc = swift.cwc.stations(basin=['Narmada', 'Tapi'], state=['Gujarat', 'Maharashtra']) # Returns a pandas Dataframe with the stations and their CWC IDs.
stations_cwc
```

```python
# We can subset the stations table as well if we are only interested in a few stations for downloading the data.
stations_cwc_subset = stations_cwc[stations_cwc['name'].isin(['BODELI', 'Bharuch'])]
stations_cwc_subset
```

```python
# Then we pass this subset table (or the full table) to the fetch function to download the data for these basins.

gdf_cwc = swift.fetch(
    stations_cwc_subset,# the subset dataframe containing the basins and their CWC IDs.
    output_dir="data_fetch_cwc", #location where the downloaded files will be saved.
    start_date="2024-04-01",
    end_date="2024-06-07",
    merge=True, # merges and returns all the downloaded files as a geodataframe.
    plot=True, # plots the station-wise time series of the downloaded data.
    quiet=False, # prints the progress of the download and any warnings or errors encountered during the download process.
    overwrite=True # overwrites any existing files in the output directory with the same name as the downloaded files.
)

# gdf_cwc will contain the merged geodataframe of the downloaded data for the specified basins and variables.
```

# Merging into a Geopackage (.gpkg)
- If we set merge = False while fetching the data, we can later use merge_only() to generate the merged geopackage.

- The input_dir in this case would be the folder where the previously downloaded csv/xlsx files are present.

```python
dg = swift.merge_only(
    mode='cwc', # mode = 'cwc' when merging cwc data 
    input_dir="data_fetch_cwc", 
    output_dir='merged_cwc') 

# the merged geopackage file will be saved in the output directory specified (merged_cwc in this case) 
# and a geodataframe containing the merged data will be returned.
dg
```

```python
dh = swift.merge_only(mode='wris', # mode = 'wris' when merging wris data 
 input_dir="data_fetch_wris",
 output_dir='merged_wris', 
 variable=['sediment', 'water_level']) # If variable is not specified, all variables in the downloaded files will be merged and returned in the geodataframe.


# the merged geopackage file will be saved in the output directory specified (merged_wris in this case) 
# and a geodataframe containing the merged data will be returned.
dh
```

# Generating Time series Plots

- If we set plot = False while fetching the data, we can later use plot_only() to generate the merged geopackage.

- The input_dir in this case would be the folder where the previously downloaded csv/xlsx files are present.

- plot_only()

```python
swift.plot_only (mode='cwc', # mode = 'cwc' when plotting cwc data
 input_dir='data_fetch_cwc',
 output_dir='plots_cwc',
    variable=['water_level'], # in cwc mode the variable argument will be ignored.
    plot_svg=True ,#svg file can also be generated
    moving_average=True # adds a moving-average overlay to the plot
    )
```

```python
swift.plot_only (mode='wris', # mode = 'cwc' when plotting cwc data
 input_dir='data_fetch_wris',
 output_dir='plots_wris',
    variable=['sediment','water_level'], # in cwc mode the variable argument will be ignored.
    plot_svg=True ,#svg file can also be generated
    moving_average=True ,
    window=30 # adds a moving-average overlay to the plot
    )
```
