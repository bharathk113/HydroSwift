# HydroSwift CLI — Beginner-Friendly Guide

This notebook explains **how to use the HydroSwift CLI step-by-step**, with clear mapping to the Python API.

👉 If `hyswift` is not found, use:

```
python -m hydroswift
```

## 1. Check Installation

Make sure CLI works:

````python
!hyswift -h
````

## 2. Core Idea (Very Important)

HydroSwift CLI is **flag-based**, not subcommand-based.

👉 You build commands like:
```
hyswift -b <basin> <dataset flags>
```

### Python equivalent:
```python
hydroswift.wris.download(variable='discharge', basin='Godavari')
```

## 3. List Available Basins

````python
!hyswift --list
````

## 4. WRIS Download — Basics

### Required flags:
- `-b` → basin
- dataset flags → what data you want

### Dataset flags:
- `-q` → discharge
- `-wl` → water level
- `-rf` → rainfall
- `-temp` → temperature
- `-rh` → humidity
- `-solar` → solar radiation
- `-sed` → sediment
- `-gwl` → groundwater


### Example 1: Discharge (Godavari)

````python
!hyswift -b 5 -q
````

Python equivalent:
```python
hydroswift.wris.download(basin=5, variable='discharge')
```

### Example 2: Multiple datasets

````python
!hyswift -b Krishna -q -rf -temp
````

Python equivalent:
```python
hydroswift.wris.download(basin='Krishna', variable=['discharge','rainfall','temperature'])
```

### Example 3: Date range

````python
!hyswift -b 6 -q --start-date 2020-01-01 --end-date 2022-01-01
````

Python equivalent:
```python
hydroswift.wris.download(basin=6, variable='discharge', start_date='2020-01-01', end_date='2022-01-01')
```

## 5. Output Control

````python
!hyswift -b 5 -q --output-dir data --format csv --overwrite
````

Python equivalent:
```python
hydroswift.wris.download(basin=5, variable='discharge', output_dir='data', format='csv', overwrite=True)
```

## 6. Plotting

````python
!hyswift -b 5 -q --plot
````

Python equivalent:
```python
hydroswift.wris.download(basin=5, variable='discharge', plot=True)
```

## 7. Merge Mode

````python
!hyswift --merge-only --input-dir output
````

Python equivalent:
```python
hydroswift.merge_only(input_dir='output')
```

## 8. Plot Only Mode

````python
!hyswift --plot-only --input-dir output
````

Python equivalent:
```python
hydroswift.plot_only(input_dir='output')
```

## 9. CWC Data (Water Level Only)

⚠️ Only water level is supported.

````python
!hyswift --cwc
````

Python equivalent:
```python
hydroswift.cwc.download(basin=["Krishna"])
```

### Specific stations

````python
!hyswift --cwc-station 032-LGDHYD
````

### Basin filter

````python
!hyswift --cwc-basin Krishna
````

## 10. Advanced Tips

- Use `--delay` to control API speed
- Use `--quiet` to suppress logs
- Combine flags for pipelines


## 11. Debugging

If CLI fails:
```
python -m hydroswift -h
```
