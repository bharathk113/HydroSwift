```python
# Invalid-input examples: these should raise clear errors
from pathlib import Path

def expect_error(label, fn):
    try:
        fn()
        print(f"❌ {label}: expected an error, but call succeeded")
    except Exception as e:
        print(f"✅ {label}: {type(e).__name__} -> {e}")

expect_error(
    "Unknown WRIS variable",
    lambda: swift.wris.basins(variable="not_a_real_variable"),
)

expect_error(
    "wris.stations() missing variable",
    lambda: swift.wris.stations("Godavari", None),
)

expect_error(
    "fetch() invalid source type",
    lambda: swift.fetch("invalid_source"),
)

missing_dir = Path("_definitely_missing_input_dir_")
expect_error(
    "plot_only() with missing input_dir",
    lambda: swift.plot_only(input_dir=str(missing_dir), output_dir="test_plot_invalid"),
)
```

```python
# Example D: expected failures for invalid combinations

def expect_error_contains(label, fn, contains):
    try:
        fn()
        print(f"❌ {label}: expected error containing {contains!r}, but call succeeded")
    except Exception as e:
        msg = str(e)
        ok = contains.lower() in msg.lower()
        prefix = "✅" if ok else "⚠️"
        print(f"{prefix} {label}: {type(e).__name__} -> {msg}")

expect_error_contains(
    "wris.download missing variable",
    lambda: swift.wris.download(basin="Krishna", variable=None),
    "variable is required",
)

expect_error_contains(
    "cwc.download basin table missing basin column",
    lambda: swift.cwc.download(basin=pd.DataFrame({"name": ["Krishna"]})),
    "must include a 'basin' column",
)

expect_error_contains(
    "fetch WRIS table missing station_code and basin/variable",
    lambda: swift.fetch(pd.DataFrame({"x": [1, 2]})),
    "infer data source",
)
```

## Smoke-only playground (fast, low-risk)
This section avoids heavy network/file operations and validates API shape + error behavior.
Run these cells first before trying download examples.

```python
# Smoke utility: pretty assertion runner
import traceback

def smoke_check(name, fn):
    try:
        result = fn()
        print(f"✅ {name}")
        return True, result
    except Exception as e:
        print(f"❌ {name}: {type(e).__name__} -> {e}")
        traceback.print_exc(limit=1)
        return False, None
```

```python
# Smoke 1: public API surface and namespace methods
required_top = ["wris", "cwc", "fetch", "merge_only", "plot_only", "help", "cli_help", "cite", "coffee"]
required_wris = ["download", "stations", "variables", "basins"]
required_cwc = ["download", "stations", "basins", "reconcile_metadata"]

ok, _ = smoke_check(
    "Top-level API surface",
    lambda: all(hasattr(swift, k) for k in required_top),
)
print("top-level ok:", ok)

print("wris missing:", [k for k in required_wris if not hasattr(swift.wris, k)])
print("cwc missing:", [k for k in required_cwc if not hasattr(swift.cwc, k)])
```

```python
# Smoke 2: metadata tables only (no downloads)
ok_wvars, wvars = smoke_check("swift.wris.variables()", lambda: swift.wris.variables())
ok_wbasins, wbasins = smoke_check("swift.wris.basins(variable=['discharge','solar'])", lambda: swift.wris.basins(variable=["discharge", "solar"]))
ok_cbasins, cbasins = smoke_check("swift.cwc.basins()", lambda: swift.cwc.basins())

if ok_wvars:
    print(wvars[["flag", "canonical_name"]].head(8))
if ok_wbasins:
    print(wbasins.head(6))
if ok_cbasins:
    print(cbasins.head(6))
```

```python
# Smoke 3: expected-invalid inputs (should fail cleanly)
def expect_fail(label, fn):
    try:
        fn()
        print(f"❌ {label}: expected failure, but call succeeded")
    except Exception as e:
        print(f"✅ {label}: {type(e).__name__} -> {e}")

expect_fail("wris.basins unknown variable", lambda: swift.wris.basins(variable="___bad___"))
expect_fail("wris.stations missing variable", lambda: swift.wris.stations(basin="Godavari", variable=None))
expect_fail("fetch invalid type", lambda: swift.fetch("not_a_table"))
expect_fail("plot_only missing input dir", lambda: swift.plot_only(input_dir="_missing_smoke_dir_", output_dir="_x_"))
```

## Formatted sample downloads playground
Small-range examples for trying combinations. Keep `overwrite=False` first; turn it on when you want a refresh.

```python
# Download setup (shared)
DL_START = "2024-01-01"
DL_END = "2024-02-03"

OUT_WRIS = "playground_output_wris"
OUT_CWC = "playground_output_cwc"
```

```python
# Combo 1: WRIS direct basin+variable download (single basin, dual variables)
wris_combo_1 = swift.wris.download(
    basin="Krishna",
    variable=["discharge", "solar"],
    start_date=DL_START,
    end_date=DL_END,
    output_dir=OUT_WRIS,
    format="csv",
    overwrite=False,
    merge=False,
    plot=False,
    quiet=False,
)
print("Combo 1 done")
```

```python
# Combo 2: WRIS station-table -> fetch (tiny slice)
wris_station_tbl = swift.wris.stations(
    basin=["Godavari"],
    variable=["sediment"],
)
wris_station_tbl_small = wris_station_tbl.head(2)

swift.fetch(
    wris_station_tbl_small,
    output_dir=OUT_WRIS,
    start_date=DL_START,
    end_date=DL_END,
    format="xlsx",
    overwrite=False,
    merge=False,
    plot=False,
    quiet=False,
)
print("Combo 2 rows used:", len(wris_station_tbl_small))
```

```python
# Combo 3: CWC basin-table filter -> fetch (tiny basin subset)
cwc_basin_tbl = swift.cwc.basins().head(2)

swift.fetch(
    cwc_basin_tbl,
    output_dir=OUT_CWC,
    start_date=DL_START,
    end_date=DL_END,
    format="csv",
    overwrite=False,
    merge=False,
    plot=False,
    quiet=False,
)
print("Combo 3 basins:", cwc_basin_tbl["basin"].tolist())
```

```python
# Combo 4: CWC direct station download + merge return
cwc_station_tbl = swift.cwc.stations(basin=["Godavari"]).head(2)
cwc_codes = cwc_station_tbl["code"].tolist()

cwc_gdf = swift.cwc.download(
    station=cwc_codes,
    start_date=DL_START,
    end_date=DL_END,
    output_dir=OUT_CWC,
    format="csv",
    overwrite=False,
    merge=True,
    plot=False,
    quiet=False,
)

print("Combo 4 station codes:", cwc_codes)
print("Returned object type:", type(cwc_gdf).__name__ if cwc_gdf is not None else None)
```

```python
# Combo 5: merge_only / plot_only on generated playground outputs (safe if dirs exist)
from pathlib import Path

for mode, in_dir, out_dir in [
    ("wris", OUT_WRIS, f"{OUT_WRIS}_post"),
    ("cwc", OUT_CWC, f"{OUT_CWC}_post"),
]:
    if Path(in_dir).exists():
        print(f"Running post-process for mode={mode} input={in_dir}")
        try:
            swift.merge_only(mode=mode, input_dir=in_dir, output_dir=out_dir)
            swift.plot_only(mode=mode, input_dir=in_dir, output_dir=out_dir)
            print("  ✅ done")
        except Exception as e:
            print(f"  ⚠️ post-process error: {type(e).__name__} -> {e}")
    else:
        print(f"Skipping mode={mode}: {in_dir} does not exist yet")
```
