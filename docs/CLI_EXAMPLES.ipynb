# SWIFT CLI Examples Notebook

This notebook collects practical command-line examples for SWIFT.
Run commands in a shell/terminal (or with `!` in Jupyter).


## 1) WRIS downloads


```python
# Discharge for Krishna basin
!swift -b Krishna -q --start-date 2024-01-01 --end-date 2024-02-01

# Multiple variables using Python-like long flags
!swift -b Godavari --discharge --rainfall --temperature
```

## 2) CWC downloads


```python
# By stations
!swift --station 040-CDJAPR 032-LGDHYD --start-date 2024-01-01 --end-date 2024-01-07

# By basin filters
!swift --cwc-basin Krishna Godavari
```

## 3) Plotting and merge helpers


```python
# Merge existing output
!swift --merge-only --input-dir output --output-dir output

# Publication-ready plots (PNG + SVG) with trend overlay
!swift --plot-only --input-dir output --plot-svg --plot-trend-window 30
```
