# SWIFT Python API Examples Notebook

Examples for WRIS/CWC API usage and table-driven fetch workflows.


```python
import swift_app as swift

swift.help()
```

## WRIS direct download


```python
gdf_wris = swift.wris.download(
    basin='Krishna',
    variable=['discharge'],
    start_date='2024-01-01',
    end_date='2024-01-07',
    merge=True,
)
gdf_wris.head() if gdf_wris is not None else None
```

## CWC direct download


```python
gdf_cwc = swift.cwc.download(
    station=['040-CDJAPR'],
    start_date='2024-01-01',
    end_date='2024-01-07',
    merge=True,
)
gdf_cwc.head() if gdf_cwc is not None else None
```

## Table-driven fetch


```python
wris_tbl = swift.wris.stations(basin='Krishna', variable='discharge').head(3)
swift.fetch(wris_tbl, start_date='2024-01-01', end_date='2024-01-07')
```

```python
cwc_tbl = swift.cwc.stations(basin='Krishna', refresh=True).head(3)
swift.fetch(cwc_tbl, start_date='2024-01-01', end_date='2024-01-07')
```
